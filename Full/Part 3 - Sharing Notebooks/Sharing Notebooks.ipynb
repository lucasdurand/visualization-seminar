## Binder
https://mybinder.org/

Binder is cool. Where with HTML we get a *static* version of our notebook, Binder will launch an ephemeral notebook server to give you the full notebook experience.

### How does it work?

It takes a repository and:

1. Builds a `Docker` image for it
    - Docker implements *containerization*, a modern programming paradigm that allows you to create "insulated" processes that once created *should* work anywhere
2. Launches a `JupyterHub` server to host your repo
    - JupyterHub is a multi-user Jupyter platform. In this case it allows us to host things externally (for a short time) and only spin it up/down when users want to access it.
3. Gives you a re-usable link to access and share.
