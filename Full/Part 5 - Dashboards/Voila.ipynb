# Widgets and Voila

A more *native* way to do dashboarding in Python is to use `ipywidgets` to build user interfaces along with `Voila` to create *apps* from Notebooks. This pairs particularly well with [Binder](https://mybinder.org/)

Check out other examples here: https://voila-gallery.org/services/gallery/

## Widgets

`ipywidgets` are an older python solution for connecting python kernel processes to front-end, especially for use in Jupyter Notebooks.

```python
import ipywidgets as widgets
```

```python
import pandas as pd
def my_plot(f=pd.np.sin, N=100, top=100, m=1):
    df = pd.DataFrame({"x":pd.np.linspace(0,top,N),str(f):f(pd.np.linspace(0,top,N)*m)})
    return df.plot(x="x",y=str(f));

my_plot(N=10);
```

## Interact

A very simple wrapper for `widgets` to allow fast interactive plots. For most real applications you will want to customize this by using the underlying [widgets](https://ipywidgets.readthedocs.io/en/latest/examples/Widget%20Basics.html) (which requires a bit of work to link together) 

```python
widgets.interactive(
    my_plot,
    f = [pd.np.sin, pd.np.cos, pd.np.cumsum],
    N = (1,1000),
    top = 100,
    m = 1.0
)
```
