# How do I share this?

It isn't immediately clear how this can be done. Do you save the file and email it to someone? Can they access the link?

We are going to focus on two main ways to share:

- HTML
- Binder.org

## HTML

As simple as saving the file. It is converted using a library called `nbconvert` and can be turned into many different forms (including PDF). What makes `HTML` great is that it is able to embed both standard images generated by `matplotlib` as well as interactive plots (that we will make later).
